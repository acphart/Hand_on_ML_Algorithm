```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell

InteractiveShell.ast_node_interactivity = 'all'
```

```python
np.random.seed(20180823)

m = 100
Theta = [[2.0], [2.0], [2.0]]

x0 = np.ones((100, 1))
x1 = np.linspace(-2, 8, 100).reshape(100, 1)
x2 = np.linspace(-2, 2, 100).reshape(100, 1)

X = np.hstack((x0, x1, x2))
y = np.dot(X, Theta) + np.random.randn(100, 1)*2

_ = plt.scatter(x1, y)
```

```python
def loss_func(X, y, theta):
    loss = np.dot(X, theta) - y
    return 1./(2*m) * np.dot(loss.T, loss)

def grad_func(X, y, theta):
    loss = np.dot(X, theta) - y
    return 1./m * np.dot(X.T, loss)
```

```python
np.random.seed(20180823)
alpha = 0.01
accuracy = 1e-5

i = 1
index = 1
c = np.array([0.8, 0.8, 0.8])

theta = np.random.randn(3, 1)
grad = grad_func(X, y, theta)
while not np.all(abs(grad) <= accuracy):
    theta = theta - alpha*grad
    grad = grad_func(X, y, theta)
    
    i = i+1
    if i%index == 0:
        _ = plt.plot(x1, np.dot(X, theta), color=c)
        c = c - [0, 0.1, 0.1]
        index = index*5

theta
_ = plt.scatter(x1, y, alpha=0.6)    
```
